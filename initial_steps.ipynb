```python
import pygame
import time

# Caminho para o seu arquivo .wav
caminho_do_arquivo = '/Users/gustavo/Faculdade/TCC/BMD-HS-Dataset/train/AR_016_sit_Mit.wav'

# Inicializa o mixer do pygame
pygame.mixer.init()

try:
    # Carrega e toca o arquivo
    pygame.mixer.music.load(caminho_do_arquivo)
    print(f"Tocando o arquivo: {caminho_do_arquivo}")
    pygame.mixer.music.play()

    # Mantém o script rodando enquanto a música estiver tocando
    while pygame.mixer.music.get_busy():
        time.sleep(1)
        
    print("Reprodução concluída!")
    
except pygame.error as e:
    print(f"Erro ao tentar reproduzir o áudio: {e}")
```
